```python
# Supervisory Learning - Logistic Regression
import seaborn as sns
import pandas as pd
import statistics as st
import matplotlib.pyplot as plt

# Read the training data from the uploaded file
data_train=pd.read_csv('/content/train_LZdllcl.csv')
data_train
```

```python
# Read the test data from the uploaded file
data_test=pd.read_csv('/content/test_2umaH9m.csv')
data_test
```

```python
# Replace the education column null values by mode
data_train.loc[data_train["education"].isna(),"education"] = st.mode(data_train["education"])
data_train.isnull().sum()
```

```python
# Replace the previous_year_rating column null values by first value
data_train.loc[data_train["previous_year_rating"].isna(),"previous_year_rating"] = data_train["previous_year_rating"][0]
data_train.isnull().sum()
```

```python
# Replace the education column null values by mode
data_test.loc[data_test["education"].isna(),"education"] = st.mode(data_test["education"])
data_test.isnull().sum()
```

```python
# Replace the previous_year_rating column null values by first value
data_test.loc[data_test["previous_year_rating"].isna(),"previous_year_rating"] = data_test["previous_year_rating"][1]
data_test.isnull().sum()
```

```python
#Define numerical columns and text columns in the train data
from sklearn.impute import SimpleImputer
num_col_train = data_train.select_dtypes(include=['number']).columns
cat_col_train = data_train.select_dtypes(exclude=["number"]).columns
```

```python
# Apply onehot encoding for minimal categories
# gender
data_train = pd.get_dummies(data_train, columns=['gender'],dtype=int,drop_first=True)
data_test = pd.get_dummies(data_test, columns=['gender'],dtype=int,drop_first=True)
```

```python
# Apply Target encoding on multiple values cloumns
#  'department', 'region', 'education', 'recruitment_channel
data_train['department']= data_train.groupby('department')['is_promoted'].transform("mean")
data_train['region']= data_train.groupby('region')['is_promoted'].transform("mean")
data_train['education']= data_train.groupby('education')['is_promoted'].transform("mean")
data_train['recruitment_channel']= data_train.groupby('recruitment_channel')['is_promoted'].transform("mean")
```

```python
data_train
```

```python
#Apply Standard Scaler on Numerical train Data
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
#Apply Standard scaling to transform data
scaler = StandardScaler()

data_train[num_col_test] = scaler.fit_transform(data_train[num_col_test])
```

```python
#APPLY Standard Scaler on Numerical test Data
data_test[num_col_test] = scaler.fit_transform(data_test[num_col_test])
```

```python
# Define features columns of trained data
data_train_array = pd.DataFrame(data_train,columns=data_train.columns)
X_tn = data_train_array.drop("is_promoted", axis=1)
X_tn = X_tn.drop("employee_id", axis=1)
# Define target column
y_tn = data_train_array["is_promoted"]
```

```python
# Define features columns of test data
data_test_array = pd.DataFrame(data_test,columns=data_test.columns)
X_tt = data_test_array
```

```python
from sklearn.linear_model import LinearRegression #first model supervised regression model

model = LinearRegression()  # model named as model
model.fit(X_tn, y_tn) # teaching the trained values - machine learning happend
```

```python
y_pred = model.predict(X_tn)
y_pred # predicted answers
```

```python
results = pd.DataFrame({
    "Actual (₹)": y_tn.values,
    "Predicted (₹)": y_pred.astype(int),
    "diff": (y_tn.values - y_pred).astype(int)

})
results
```

```python
from sklearn.metrics import r2_score, mean_absolute_error # regression score and mean absolute error
# Regression metrics 99.63% accuracy
r2 = r2_score(y_tn, y_pred)
mae = mean_absolute_error(y_tn, y_pred)

print(f"\nR² Score: {r2:.4f}")
print(f"Mean Absolute Error: ₹{int(mae):,}")
```
