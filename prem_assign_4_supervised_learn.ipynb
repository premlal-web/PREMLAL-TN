```python
# Apply onehot encoding for minimal categories
# marital, default, housing,	loan, contact
data = pd.get_dummies(data, columns=['marital'],dtype=int,drop_first=True)
data = pd.get_dummies(data, columns=['default'],dtype=int,drop_first=True)
data = pd.get_dummies(data, columns=['housing'],dtype=int,drop_first=True)
data = pd.get_dummies(data, columns=['loan'],dtype=int,drop_first=True)
data = pd.get_dummies(data, columns=['contact'],dtype=int,drop_first=True)
```

```python
data
```

```python
# Apply Target encoding since there multiple values in the below feature cloumns
#  job, education, month, day_of_week,	poutcome
data['job']= data.groupby('job')['nr.employed'].transform("mean")

data['education'] = data.groupby('education')['nr.employed'].transform("mean")

data['month'] = data.groupby('month')['nr.employed'].transform("mean")

data['day_of_week'] = data.groupby('day_of_week')['nr.employed'].transform("mean")

data['poutcome'] = data.groupby('poutcome')['nr.employed'].transform("mean")
```

```python
data['y'] = data['y'].replace({'yes': 1, 'no': 0})
```

```python
data
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
#Apply Standard scaling to transform data
scaler = StandardScaler()

data[num_col] = scaler.fit_transform(data[num_col])
```

```python
# MODEL BUILDING USING LOGESTIC REGRESSSION
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression

param_grid = {
    "C": [0.01, 0.1, 1],          # if smaller then simpler models, Larger values more complex, can overfit
    "solver": ["lbfgs", "liblinear"], # weight optimization lbfgs for large data, other for small data
    "penalty": ["l2","elastic","l1"]  # avoid over fitting
}
# 3 x 2 x 3 = 18 combinations possible

# GridSearchCV to check all the combinations
grid = GridSearchCV(LogisticRegression(max_iter=100), param_grid, cv=5, scoring="accuracy", n_jobs=-1)
grid.fit(X_train, y_train)

best_log_reg = grid.best_estimator_

# Train & Test Accuracy after tuning
train_acc_tuned_logreg = accuracy_score(y_train, best_log_reg.predict(X_train))
test_acc_tuned_logreg = accuracy_score(y_test, best_log_reg.predict(X_test))

# PRINT OUTPUT
print("\nLogistic Regression (Tuned)")
print("Best Params:", grid.best_params_)
print("Train Accuracy:", train_acc_tuned_logreg)
print("Test Accuracy:", test_acc_tuned_logreg)
```
