```python
#--------- HISTOGRAM OF MARKS -----------------
import seaborn as sns # import library
import pandas as pd
df=pd.read_excel("student_data.xlsx")

sns.histplot(x=df["Marks"],data=df)
```

```python
# DIST PLOT OF STUDY HOURS -----------
sns.distplot(df["StudyHours"])
```

```python
#------- COUNT PLOT OF GENDER ------------
sns.countplot(x=df["Gender"])
```

```python
#--------- BAR PLOT OF MARKS BY STUDENT ------------------
import matplotlib.pyplot as plt
x = df["StudentID"]
y = df["Marks"]

plt.bar(x,y, color='b')
plt.title("Student Marks Graph")
plt.xlabel("Students")
plt.ylabel("Marks")
```

```python
#---- BOXPLOT OF MARKS GROUP BY CLASS------------
sns.boxplot(x=df["Class"],y=df["Marks"])
```

```python
#----- SCATTER PLOT OF STDUYHOURS AND MARKS ----
sh = df["StudyHours"]
mk = df["Marks"]
sns.scatterplot(x=sh,y=mk)
```

```python
#----- STRIP PLOT OF MARKS AND SUBJECT USING GENDER AS hue ----
sns.stripplot(x='Marks',y='Subject',data=df, hue="Gender",palette="rainbow")
```

```python
#----- CATEGORICAL PLOT OF MARKS AND SUBJECT ACROSS CLASS ----
sns.catplot(x='Marks',y='Subject',data=df, hue="Class")
```

```python
#----- LM PLOT OF STUDY HOURS AND MARKS ----
sns.lmplot(x='StudyHours',y='Marks',data=df)
```

```python
#----- VIOLIN PLOT OF STUDY HOURS AND MARKS ----
sns.violinplot(x=df["StudyHours"],y=df["Marks"])
```
