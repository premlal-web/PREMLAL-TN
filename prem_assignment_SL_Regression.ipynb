```python
# Apply onehot encoding for minimal categories
data = pd.get_dummies(data, columns=['Transmission'],dtype=int,drop_first=True)
data = pd.get_dummies(data, columns=['Seller Type'],dtype=int,drop_first=True)
data = pd.get_dummies(data, columns=['Drivetrain'],dtype=int,drop_first=True)
```

```python
#Apply Target encoding since there multiple values in the below feature cloumns
data['Make'] = data['Make'].map(data.groupby('Make')['Price'].mean())
data['Model'] = data['Model'].map(data.groupby('Model')['Price'].mean())
data['Owner'] = data['Owner'].map(data.groupby('Owner')['Price'].mean())
data['Fuel Type'] = data['Fuel Type'].map(data.groupby('Fuel Type')['Price'].mean())
```

```python
#Apply Target encoding since there multiple values in the below feature cloumns
data['Location'] = data['Location'].map(data.groupby('Location')['Price'].mean())
data['Color'] = data['Color'].map(data.groupby('Color')['Price'].mean())
data['Engine'] = data['Engine'].map(data.groupby('Engine')['Price'].mean())
data['Max Power'] = data['Max Power'].map(data.groupby('Max Power')['Price'].mean())
data['Max Torque'] = data['Max Torque'].map(data.groupby('Max Torque')['Price'].mean())
data
```

```python
# check null value
data.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
#Define numerical columns and text columns
num_col = data.select_dtypes(include=['number']).columns
cat_col = data.select_dtypes(exclude=["number"]).columns

# Removes null values by median
num_imputer = SimpleImputer(strategy="median")
data[num_col] = num_imputer.fit_transform(data[num_col])
data.isnull().sum()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
#Apply Standard scaling to transform data
scaler = StandardScaler()

data[num_col] = scaler.fit_transform(data[num_col])
```

```python
X = data.drop("Price", axis=1) #define feature columns
y = data["Price"] #define target column

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)
```

```python
from sklearn.linear_model import LinearRegression #first model supervised regression model

model = LinearRegression()  # model named as model
model.fit(X_train, y_train) # teaching the trained values - machine learning happend
```

```python
y_pred = model.predict(X_test) # predict a targe model using test data
y_test
y_pred # predicted answers
```

```python
results = pd.DataFrame({
    "Actual (₹)": y_test.values,
    "Predicted (₹)": y_pred.astype(int),
    "diff": (y_test.values - y_pred).astype(int)

})
results # define actual and predict price
```

```python
from sklearn.metrics import r2_score, mean_absolute_error # regression score and mean absolute error
# Regression metrics 99.63% accuracy
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)

print(f"\nR² Score: {r2:.4f}")
print(f"Mean Absolute Error: ₹{int(mae):,}")
```
